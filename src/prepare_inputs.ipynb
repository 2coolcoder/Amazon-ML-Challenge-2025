```python
import re
def find_values(ds, nan_value_ids):
    k=0
    # If both unit and value are missing, try to extract from Item Name
    for i in ds.loc[nan_value_ids].index:

        # Try to extract unit and value from item_line
        # Pattern example: "7 oz", "2.5 lb", "12 count", etc.
        pattern = r'(\d+\.?\d*)\s?(oz|ounce[s]?|fl\.?\s?oz|fluid ounce[s]?|lb|pound[s]?|g|gram[s]?|gm|gr|ml|milliliter[s]?|mililitro|ct|count|each|pack[s]?|box|bottle[s]?|case|jar|can|pouch|capsule|unit[s]?)'

        for col in ['Name', 'Bullet Points', 'Description']:
            found = re.search(pattern, ds.loc[i, col])
            if found:
                value = found.group(1)
                unit = found.group(2).replace('.', '').strip()  # Remove dots in e.g. fl. oz
                # print(value, unit)
                ds.loc[i,'Value'] = float(value)
                ds.loc[i,'Unit'] = unit
                k+=1
                break
    print(k, "values found")
find_values(train, nan_value_ids)
```

```python
def normalize_units(df, value_col='Value', unit_col='Unit'):
    # --- Mappings ---
    weight_units = {
        'ounce': 28.3495, 'oz': 28.3495, 'ounces': 28.3495, 'oz.': 28.3495,
        'pound': 453.592, 'lb': 453.592, 'lbs': 453.592, 'kg': 1000, 'gram': 1, 'g': 1, 'grams': 1
    }

    volume_units = {
        'fl oz': 29.5735, 'fluid ounce': 29.5735, 'fluid ounces': 29.5735,
        'liter': 1000, 'litre': 1000, 'liters': 1000, 'milliliter': 1, 'millilitre': 1, 'ml': 1
    }

    count_units = {
        'count': 1, 'each': 1, 'ct': 1, 'unit': 1, 'units': 1, 'pack': 1, 'packs': 1,
        'bottle': 1, 'bag': 1, 'box': 1, 'can': 1, 'piece': 1
    }

    area_units = {
        'sq ft': 1, 'square foot': 1
    }

    # --- Cleaning function for unit text ---
    def clean_unit(u):
        if pd.isna(u):
            return None
        u = str(u).strip().lower()
        u = u.replace('.', '').replace('(', '').replace(')', '')
        u = u.replace('/', ' ').replace('-', ' ').strip()
        return u

    # --- Apply cleaning ---
    df = df.copy()
    df[unit_col] = df[unit_col].apply(clean_unit)

    normalized_values = []
    normalized_units = []

    # --- Conversion loop ---
    for val, unit in zip(df[value_col], df[unit_col]):
        if unit in weight_units:
            base_val = val * weight_units[unit]
            base_unit = 'g'
        elif unit in volume_units:
            base_val = val * volume_units[unit]
            base_unit = 'ml'
        elif unit in count_units:
            base_val = val  # no conversion needed
            base_unit = 'count'
        elif unit in area_units:
            base_val = val  # keep same
            base_unit = 'sq ft'
        else:
            base_val = val
            base_unit = 'Other'  # unknown
        normalized_values.append(base_val)
        normalized_units.append(base_unit)

    df['NormalizedValue'] = normalized_values
    df['NormalizedUnit'] = normalized_units
    return df
```
